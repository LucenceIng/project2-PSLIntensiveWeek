```python
#on vérifie l'alignement
collate(json,output="html",segmentation=False)
```

```python
#on réalise la collation sous forme de table et on stocke le résultat
table = collate(json, output="table",segmentation=False)
```

```python
#on va définir notre fonction d'export qui va produire un XML, qui va reprendre les données de lemmes et de textes

def notre_export_xml(table):
    readings = []
    for column in table.columns:
        #définition d'un élément app
        app = et.Element('app')
        for key, value in sorted(column.tokens_per_witness.items()):
            #pour chaque élément de la colonne, création d'un rdg
            child = et.Element('rdg')
            #création d'un attribut wit
            child.attrib['wit'] = "#" + key
            #création d'un attribut lemma
            for item in value:
                child.attrib['lemma']= item.token_data["n"]
             #on retourne la valeur du texte    
            child.text = "".join(str(item.token_data["t"]) for item in value)
            app.append(child)
        result = et.tostring(app, encoding="unicode")
        readings.append(result)
    return "".join(readings)
```

```python
#on applique la fonction pour voir ce que ça donne
notre_export_xml(table)
```

```python
# on stocke l'export
resultatColl = notre_export_xml(table)

#il faut créer un document valide, donc créer une balise racine
resultatCollValide = "<text>" + resultatColl + "</text>"
print(resultatCollValide)
```

```python
#on écrit le résultat (une simple chaîne de caractères) dans notre document XML
with open("resultat_collation_json_avecLemmes.xml", "w") as text_file:
    text_file.write(resultatCollValide)
```
