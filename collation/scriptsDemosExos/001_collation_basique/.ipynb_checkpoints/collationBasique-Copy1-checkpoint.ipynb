```python
#importer collatex
from collatex import *
```

```python
#on crée l'objet collation de classe Collation()
collation = Collation()

#on ajoute deux témoins : on a besoin de l'identifiant du témoin et du texte
collation.add_plain_witness("A", "En l’an mil IIIC XLIX alerent les penitanciers qui yssirent premierement d’Alemaigne. Et furent gens qui faisoient penitances publiques et se batoient d’escorgees a bourdons et a aguillons de fer, tant qu’ilz desciroient leurs dos et leurs espaulles, et chantoient chançons moult piteuses de la nativité et souffrance de Nostre Signeur, et ne pouoient par leur ordonnance gesir que une nuit en une bonne ville et se partoient d’une ville par compaignie tant du plus comme du mains, et alerent ainsi par le paÿs faisans leur penance XXXIII jours et demy, autant que Jhesu Crist ala par terre, puis retournoient en leurs lieux, si fu ceste chose commencee par grant humilité et pour prier Nostre Seigneur qu’il voulsist refraindre son ire et cesser ses verges. Car en ce temps par tout le monde generalment, une maladie couru que on appelloit epidimie, dont bien la tierce partie mouru.")
collation.add_plain_witness("B", "En l’an de grasce Nostre Segneur M CCC et XLIX alerent ly penant et issirent premierement d’Allemagne. Et furent gens qui fasoient penitances publiques et se battoient d’escorgies a bourdons et a aguillons de fier, tant qu’il desciroient leur dos et leurs espaulles. Et chantoient canchons moult piteuse de le nativité et souffranche Nostre Signeur. Et ne pooient par leur ordenanche gesir que une nuit en une bonne bonne ville et se partoient d’une ville par compagnie tant dou plus tant dou mains. Et aloient ensy par le païs faisant leur penitance XXXIII jours et demi, otant que Jhesus Cris ala par terre d’ans, et puis retournoient en leurz lieus. Sy fu ceste cose comnmenchie par grant humilité et pour priier a Nostre Signeur qu’il vosist entendre refraindre son ire et cesser ses verghes. Car en che temps par tout le monde generalment une maladie que on claimme epidimie, couroit, dont bien la tierche partie dou monde morut.")
```

```python
#on collationne
alignment_table = collate(collation)
```

```python
#par défaut, table
print(alignment_table)
```

```python
#ajout de l'option : segmentation=False pour aligner mot à mot
alignment_table = collate(collation,segmentation=False)

print(alignment_table)
```

```python
#sortie HTML
collate(collation, output="html", layout="vertical",segmentation=False)
```

```python
#HTML colloré
aEnr = collate(collation, output="html", segmentation=False)
```

```python
type(aEnr)
```

```python
aEnr = collate(collation, output="table", segmentation=False)
```

```python
aEnrStr = str(aEnr)
```

```python
#enregistrement du résultat dans un fichier qu'on crée
with open("resultat_collation_basique", "w") as text_file:
    text_file.write(aEnrStr)
```

```python
# sortie XML
collate(collation,output="xml",segmentation=False)
```

```python
# sortie TEI
collate(collation, output="tei", segmentation=False, near_match=True, indent=True)
```

```python
## le fonctionnement de l'alignement

#nouvel objet collation
coll = Collation()
#on ajoute les témoins
coll.add_plain_witness("A", "qui soient plus boenes que beles")
#fausse version de B
coll.add_plain_witness("B", "qui soient plus bonnes que tres belles")
```

```python
#table
print(collate(coll, segmentation=False))
```

```python
#utilisation de la fonction near_match dans la table
print(collate(coll, segmentation=False,near_match=True))
```

```python
## ajouter un témoin

#nouvel objet collation
collation2 = Collation()
```

```python
#on modifie le témoin A
collation2.add_plain_witness("A", "Mais au povre home doit l’en doner tex choses qui soient plus boenes que beles, et plus porfitables que plaisanz.")
#on garde B
collation2.add_plain_witness("B", "mais aux povres hommes doit on telles choses donner qui soient plus bonnes que belles.")
```

```python
#on ajoute le témoin C
collation2.add_plain_witness("C", "Mais au povre home doit l’en douner tex choses qui soient plus boines que beles et plus profitables que plaisans.")
```

```python
#produire en mode table le résultat de la collation à trois témoins
collate(collation2, output="html",near_match=True,segmentation=False)
```

```python
#sortie HTML2
collate(collation2, output="html2",near_match=True,segmentation=False)
```

```python
##enregistrer le résultat dans un fichier

# production sur les trois témoins de la collation avec le format XML
collXML = collate(collation2,output="xml",segmentation=False,indent=True)

#enregistrement du résultat dans un fichier qu'on crée
with open("resultat_collation_basique.xml", "w") as text_file:
    text_file.write(collXML)
```
