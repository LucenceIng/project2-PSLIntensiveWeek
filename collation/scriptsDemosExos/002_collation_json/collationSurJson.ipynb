```python
#importer collatex
from collatex import *
#importer json
import json
```

```python
#on crée le json 
json_input = """{
  "witnesses" : [
    {
      "id" : "A",
      "content" : "Mais au povre home doit l’en doner tex choses qui soient plus boenes que beles"
    },
    {
      "id" : "B",
      "content" : "mais aux povres hommes doit on telles choses donner qui soient plus bonnes que belles"
    },
    {
      "id" : "C",
      "content" : "Mais au povre home doit l’en douner tex choses qui soient plus boines que beles"
    }
  ]
}"""
#on le charge en tant que json
doc = json.loads(json_input)
```

```python
#vérification
doc
```

```python
#collation
print(collate(doc))
```

```python
#collation avec segmentation
print(collate(doc,segmentation=False))
```
